```python
import os
import shutil
import random

# =====================================================================
# CONFIGURATION LAYER (UPDATED FOR YOUR D: DRIVE)
# =====================================================================
# ═ FIXED: Pointed directly to your active absolute system directories ═
SOURCE_DIR = "D:\\dental" 
TARGET_DIR = "D:\\dental_dataset"
SPLIT_RATIO = 0.8  # 80% Train, 20% Validation

CLASSES = [
    "calculus", 
    "data_caries", 
    "gingivitis", 
    "hypodontia", 
    "tooth_discoloration", 
    "ulcer"
]

# =====================================================================
# RUN DISTRIBUTION ENGINE
# =====================================================================
print("📁 Initializing Dataset Structural Split Engine...")

if not os.path.exists(SOURCE_DIR):
    print(f"❌ Error: Cannot find the source directory '{SOURCE_DIR}'")
    print("👉 Make sure your folder path on the D: drive matches exactly!")
else:
    for cls in CLASSES:
        source_cls_dir = os.path.join(SOURCE_DIR, cls)
        if not os.path.exists(source_cls_dir):
            print(f"⚠️ Warning: Folder '{cls}' not found inside source, skipping.")
            continue
            
        all_images = [f for f in os.listdir(source_cls_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        
        random.seed(42)
        random.shuffle(all_images)
        
        split_idx = int(len(all_images) * SPLIT_RATIO)
        train_images = all_images[:split_idx]
        val_images = all_images[split_idx:]
        
        train_target_dir = os.path.join(TARGET_DIR, "train", cls)
        val_target_dir = os.path.join(TARGET_DIR, "val", cls)
        
        os.makedirs(train_target_dir, exist_ok=True)
        os.makedirs(val_target_dir, exist_ok=True)
        
        for img in train_images:
            shutil.copy(os.path.join(source_cls_dir, img), os.path.join(train_target_dir, img))
        for img in val_images:
            shutil.copy(os.path.join(source_cls_dir, img), os.path.join(val_target_dir, img))
            
        print(f"✨ Class '{cls}' split complete -> Train: {len(train_images)} | Val: {len(val_images)}")
        
    print(f"\n🚀 Execution Success! Processed dataset saved to: '{TARGET_DIR}'")
```
